#### 2. BCCI India's International Fixture <a name = '2'></a>

```python
driver = webdriver.Chrome('./chromedriver.exe')
driver.get('https://www.bcci.tv/')
```

```python
fixtureLink = driver.find_element_by_xpath('/html/body/div[3]/div/div[2]/div[2]/nav/ul/li[1]/div[2]/div/ul/li[1]/a').get_attribute('href')
```

```python
driver.get(fixtureLink)
```

```python
matchInfo = driver.find_element_by_xpath('/html/body/div[4]/div/div/div[2]/section/div/div/a[1]')
matchInfo.text.split('\n')
```

```python
def matchInfo(number):
    matchInfo = driver.find_element_by_xpath(f'/html/body/div[4]/div/div/div[2]/section/div/div/a[{number}]').text.split('\n')
    title = matchInfo[6]
    series = matchInfo[5]
    place = matchInfo[7]
    date = ' '.join(matchInfo[1:3])
    time = matchInfo[3]
    
    return title, series, place, date, time
```

```python
title = list()
series = list()
place = list()
date = list()
time = list()

for i in range(1, 20):
    print(f'Getting Match {i} Info')
    try:
        t, s, p, d, ti = matchInfo(i)
        title.append(t)
        series.append(s)
        place.append(p)
        date.append(d)
        time.append(ti)
    except NoSuchElementException:
        break
```

```python
matchDf = pd.DataFrame()

matchDf['Title'] = title
matchDf['Series'] = series
matchDf['Date'] = date
matchDf['Time'] = time
matchDf['Place'] = place

matchDf
```

```python
driver.close()
```

#### 5. Trending GitHub Repository <a name = '5'></a>

```python
driver = webdriver.Chrome('./chromedriver.exe')
driver.get('https://github.com/')
```

```python
trendingLink = driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul[2]/li[3]/a').get_attribute('href')
driver.get(trendingLink)
```

```python
trendingRepo = driver.find_elements_by_class_name('Box-row')
trendingRepo[0].text.split('\n')
```

```python
trendingRepo[1].text.split('\n')
```

```python
trendingRepo[6].text.split('\n')
```

```python
repoName = []
repoDesc = []
contributorCount = []
language = []

for i in range(0, len(trendingRepo)):
    if len(trendingRepo[i].text.split('\n')) == 5:
        repoName.append(trendingRepo[i].text.split('\n')[1])
        repoDesc.append(trendingRepo[i].text.split('\n')[2])
        if len(trendingRepo[i].text.split('\n')[3].split()) == 4:
            language.append('Language Info Not Available')
            contributorCount.append(trendingRepo[i].text.split('\n')[3].split()[1])
        elif len(trendingRepo[i].text.split('\n')[3].split()) == 5:
            language.append(trendingRepo[i].text.split('\n')[3].split()[0])
            contributorCount.append(trendingRepo[i].text.split('\n')[3].split()[2])
            
    elif len(trendingRepo[i].text.split('\n')) == 4:
        repoName.append(trendingRepo[i].text.split('\n')[1])
        repoDesc.append('N/A')
        language.append(trendingRepo[i].text.split('\n')[2].split()[0])
        contributorCount.append(trendingRepo[i].text.split('\n')[2].split()[2])
```

```python
repoDf = pd.DataFrame()

repoDf['Title'] = repoName
repoDf['Description'] = repoDesc
repoDf['Contributors Count'] = contributorCount
repoDf['Language Used'] = language

repoDf
```

```python
driver.close()
```

#### 6. Top 100 songs on Billboard <a name ='6'></a>

```python
driver = webdriver.Chrome('./chromedriver.exe')
driver.get('https://www.billboard.com/')
```

```python
driver.find_element_by_xpath('//*[@id="root"]/div[2]/div[2]/nav/ul/li[3]/a').click()
```

```python
chartList = driver.find_element_by_xpath('//*[@id="charts"]/div/div[7]/div/ol/li[11]')
chartList.text.split('\n')
```

```python
def songScrapper(number):
    songInfo = driver.find_element_by_xpath(f'//*[@id="charts"]/div/div[7]/div/ol/li[{number}]').text.split('\n')
    return songInfo[2], songInfo[3], songInfo[4], songInfo[5], songInfo[6]
```

```python
song = list()
artist = list()
lastWeekRank = list()
peakRank = list()
weeksOnBoard = list()

for i in range(1,110):
    print(f'Getting song {i}')
    try:
        s,a, lwr, pr, woc = songScrapper(i)
        song.append(s)
        artist.append(a)
        lastWeekRank.append(lwr)
        peakRank.append(pr)
        weeksOnBoard.append(woc)
        print(f'Successfully Collected Song {i} Info')
    except IndexError:
        print('Advertisement. Skipping!')
        continue
```

```python
rankDf = pd.DataFrame(index=[i for i in range(1, 101)])

rankDf['Song'] = song
rankDf['Artist'] = artist
rankDf['Last Week Rank'] = lastWeekRank
rankDf['Peak Rank'] = peakRank
rankDf['Weeks On Board'] = weeksOnBoard

rankDf
```

```python
driver.close()
```

#### 9. Most watched TV series <a name = '9'></a>

```python
driver = webdriver.Chrome('./chromedriver.exe')
driver.get('https://www.imdb.com/list/ls095964455/')
```

```python
seriesElement = driver.find_elements_by_class_name('lister-item-content')
seriesElement[0].text.split('\n')
```

```python
name = list()
yearSpan = list()
genre = list()
runTime = list()
ratings = list()
votes = list()

for i in range(0, len(seriesElement)):
    series = seriesElement[i].text.split('\n')
    if i == 1:
        yearSpan.append(series[0].split()[-2:-1])
    else:
        yearSpan.append(series[0].split()[-1])
    name1 = series[0].split()
    name.append(' '.join(name1[1:len(name1)-1]))
    genre.append(series[1].split(' | ')[2])
    runTime.append(series[1].split(' | ')[1])
    ratings.append(series[2])
    votes.append(series[6].split()[1])
```

```python
seriesDf = pd.DataFrame()

seriesDf['Name'] = name
seriesDf['Genre'] = genre
seriesDf['Ratings'] = ratings
seriesDf['Runtime'] = runTime
seriesDf['Year Span'] = yearSpan
seriesDf['Votes'] = votes

seriesDf
```

```python
driver.close()
```
